```python
import pandas as pd

data = {
    "Region": ["South", "North", "East", "West", "South", "North"],
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet"],
    "Sales": [50000, 30000, 45000, 25000, 20000, 35000]
}

df = pd.DataFrame(data)

print(df)
```

```python
south_sales = df[df["Region"] == "South"]

print(south_sales)
```

```python
high_sales = df[df["Sales"] > 30000]

print(high_sales)
```

```python
sorted_df = df.sort_values("Sales")

print(sorted_df)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()

print(product_sales)
```

```python
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun"]

sales = [10000,12000,15000,13000,18000,20000]

plt.figure(figsize=(18,5)) # Adjusted figsize to accommodate three plots

# Line Chart
plt.subplot(1,3,1) # Changed to 1 row, 3 columns, position 1

plt.plot(months,
         sales,
         marker="o")

plt.title("Sales Trend")

plt.xlabel("Months")

plt.ylabel("Sales")

# Bar Chart
plt.subplot(1,3,2) # Changed to 1 row, 3 columns, position 2

plt.bar(months,
        sales,
        color="orange")

plt.title("Sales Comparison")

plt.xlabel("Months")

plt.ylabel("Sales")

#Pie chart
plt.subplot(1,3,3) # Changed to 2 row, 3 columns, position 3

expenses = [5000,3000,2000]

categories = ["Food","Travel","Shopping"]

plt.pie(expenses,
        labels=categories,
        autopct="%1.1f%%", colors = ["pink","yellow", "green"])

plt.title("Expense Distribution")


plt.tight_layout()

plt.show()
```
